```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

#loading the dataset and setting to 'df' variable
emails = pd.read_csv('emails.csv')
emails.head(4)
```

```python
emails.info()
```

DATA WRANGLING

```python
#number of rows and columns
emails.shape
fid=len(emails.columns)
ed=len(emails.index)
print(f'There are {ed} rows and {fid} columns in the dataset')

emails.shape
```

```python
#duplicate and null columns
dub = emails.duplicated().sum()
print(f'There are {dub} duplicate')

cols = emails.isna().any(axis=1).sum()
rows = emails.isna().any(axis=0).sum()
print(f'There are {cols} and {rows} null values in the columns and rows respectively')
```

```python
emails.describe()
```

```python
#converting "prediction" column to a categorical data type
emails['Prediction'] = emails['Prediction'].astype('category')
```

```python
emails.groupby('Prediction').count()
```

DATA VISUALIZATION

```python
plt.figure(figsize=(4,4))
sns.set_context('poster')
sns.set_style('whitegrid')
sns.countplot(x='Prediction', data=emails, palette='seismic').set(title='SPAM OR NO SPAM PREDICTION COUNT PLOT')
plt.xticks([0,1],['no spam','Spam'])
sns.despine()
```

```python
classSum=emails.sum(axis=0) 
emails["sum"] =emails.sum(axis=1)
emails =emails.append(classSum,ignore_index=True)
rowSum=emails.sum(axis=1)
emails.at[5172,'index']='5172'
emails
```

```python
print('THE TOP TEN COLUMNS')
top_columns = emails.select_dtypes('int').sum().sort_values(ascending=False).head(11)
print(top_columns)


top_rows = emails.sum(axis=1).nlargest(11)
print(' ')
print('THE TOP TEN ROWS')
print(top_rows)
```

```python
top_columns.pop('sum')

top_columns.plot(kind='bar',title='TOP 10 COLUMN WITH HIGHEST OCCURENCE')
plt.xlabel('Column')
plt.ylabel('Sum of counts')
```

```python
top_rows.pop(5172)
top_rows.plot(kind='bar',title='TOP 10 ROWS WITH HIGHEST OCCURENCE')
plt.xlabel('Row number')
plt.ylabel('Sum of counts')
```
